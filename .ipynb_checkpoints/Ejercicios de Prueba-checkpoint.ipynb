```python
lista = [1,2,3]
contador = 0
for i in lista:
    print(i)
for i in range(0,11):
    print(i)
for i in range(10, -1, -1):
    print(i)
while contador <= 10:
    print(contador) 
    contador += 1 
```

```python
# Definicion de funciones
def cuadrado(n):
    return n*n
def factorial(n):
    if n <=1:
        return n 
    else:
        return n * factorial(n-1)
print(cuadrado(5))
print(factorial(5))
```
